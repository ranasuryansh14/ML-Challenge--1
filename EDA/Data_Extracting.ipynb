#### Importing Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "Parkinsons disease raw data.txt"

with open(file_path, "r") as file:
    lines = file.readlines()

# Display first few lines to inspect the format
for i in range(25):
    print(lines[i])
```

### Coverting Txt File Dataset to CSV

```python
data = []
row = []

for line in lines:
    line = line.strip()  # Remove leading/trailing spaces and newlines
    
    if line.startswith("phon_"):  # Start of a new record
        if row:  # Append previous record before starting a new one
            data.append(row)
        row = [line]  # Start new record
    
    else:  # Numerical values belong to the current record
        row.append(float(line))

data.append(row)
# Convert to DataFrame
df = pd.DataFrame(data)

# Rename columns
df.columns = ["ID"] + [f"feature_{i}" for i in range(1, len(df.columns))]

df.head(30)
```

## Dataset Details: 
This dataset is composed of a range of biomedical voice measurements. There are 
totally 195 observations and 24 features.  
If we number the features from 0, then the 17th feature is your target column.  
1 --> Parkinson’s;  0 --> Healthy 

```python
target_column = df.columns[17]
print("The target column is:", target_column)
```

```python
count = df[target_column].value_counts()
count
```

```python
df.rename(columns={df.columns[17]: "target"}, inplace=True)
df["target"] = df["target"].astype(int)  
```

```python
count_ones = df["target"].sum()  # Since 1s sum up to their count
count_zeros = (df["target"] == 0).sum()

print(f"Number of 1s (Parkinson's cases): {count_ones}")
print(f"Number of 0s (Healthy cases): {count_zeros}")
```

```python
df["target"].value_counts().plot(kind="bar", color=["green", "red"])
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Distribution of Parkinson’s Disease Cases")
plt.xticks(ticks=[0, 1], labels=["Healthy (0)", "Parkinson's (1)"])
plt.show()
```
